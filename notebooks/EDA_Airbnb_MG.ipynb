

*   Table : listing_paris



```python
#charger la table : listing_paris : Détail de chaque annonce (prix, type, localisation…)
!wget https://data.insideairbnb.com/italy/puglia/puglia/2025-09-29/visualisations/listings.csv
df_listing_paris = pd.read_csv("/content/listings.csv")
df_listing_paris.head()

#Italie > Paris
```

```python
#vérifier types de colonnes et non-null
df_listing_paris.info()
df_listing_paris.describe()
```

```python
#vérifier le nombre de doublons
df_listing_paris.duplicated().sum()
```

```python
#focus non-null
df_listing_paris.isnull().sum()
```

```python
#remplacer Null colonne host_name par "Unknown"
df_listing_paris['host_name'] = df_listing_paris['host_name'].fillna("Unknown")
```

```python
#gérer les Null colonne Price
##supprimer les lignes
##df_listing_paris = df_listing_paris.dropna(subset=['price'])
## ou remplacer par médiane
##df_listing_paris['price'] = df_listing_paris['price'].fillna(df_listing_paris['price'].median())
```

```python
#Gestion des Null colonne License
df_listing_paris['license'] = df_listing_paris['license'].fillna("No license")
```

```python
#Gestion des Null colonnes reviews_per_month & last_review
df_listing_paris['reviews_per_month'] = df_listing_paris['reviews_per_month'].fillna(0)
df_listing_paris['last_review'] = df_listing_paris['last_review'].fillna("No review")
```

```python
#vérification data
df_listing_paris.isnull().sum()
```
